Using iterrows right now for convenience, but it is very slow. 

```python
all_days=pd.DataFrame()
for day in range(1,32):
    day_df=data[data['day']==day]
    index=0
    prev_row=day_df.iloc[index]
    days=pd.DataFrame()
    count=0
    for index,row in day_df.iterrows():
        if row['hour']>=prev_row['hour']:
            count+=1
        else:
            newrow=pd.concat([row, pd.Series(count)], axis=1)
            #print(row)
            days=pd.concat([days,newrow])
            count=0
        prev_row=row
    all_days=pd.concat([all_days, days])
```
